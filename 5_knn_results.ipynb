```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from acnh import plot_trends
```

```python
mafia_data = pd.read_csv('data/turnip_mafia_data.csv')
mafia_data = pd.melt(
    mafia_data, 
    id_vars=['trend_id', 'pattern_knn'], 
    value_vars=mafia_data.columns[1:13],
    var_name='time',
    value_name='price'
)
```

```python
plot_trends(
    data=mafia_data, 
    subset_column='pattern_knn', 
    hue='trend_id', 
    title='Predicted KNN Patterns for Turnip Mafia Data'
)
```
